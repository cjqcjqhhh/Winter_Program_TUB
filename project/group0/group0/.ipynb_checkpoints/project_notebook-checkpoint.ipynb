## Main objective:
---
Imagine we are a grocery store owner, and we are trying to maximize the purchases of our customers per visit. 

A first strategy that comes to our mind is placing products next to each other that are usually bought together.

Since we have succesfully completed a Data Science task in the past we immediately realize that this problem can be formulated as a recommendation task.


The machine learning part has the following goal:


Essentially we will try to predict the last item of a customers purchase list, given all the other items that he has already in his shopping basket. Those predictions are a helpful first heuristic for the placement of certain products in our grocery store. 

Thus we start collecting the purchase histories of past customers and start writing down the following steps needed, to build our recommendation pipeline:


### Plan of attack:
1. Load the customer purchase data, located in 'data/training_data.csv', 'data/training_labels.csv'

    - Find the pair of products, that are bought together the most
    - How many customers purchased all the products 
    - Which product was the least purchased ?


4. Transform it into a Machine learning-classifier digestable format:
    - Machine learning algorithms consume data, that has a unified format!
    - For example it should look like that:
    
    
    | feature 1(e.g. product/grocery): | feature 2: | ... | feature N: |
    | "apple"                          | "banana"   | ... | mango      |

    - Item-predictions should be in the original string format(=item name)



6. Save the predictions for the test set in a csv-file

## Class DataProcess   
    class DataProcess(train_x, train_y)  

train_x is the raw datapoint (lists of products) and train_y is the raw label (next product to buy).

**<font>This class transform raw data into machine learning-classifier digestable format, including NaN dropping, one-hot encoding, train-data split and data ouput.</font>**

### Methods
    __init__(self, df_items, df_labels)

Initialize self.

**<font>Parameters:</font>**

df_items: DataFrame of raw training data

df_labels: DataFrame of raw training labels    
    
    data_preproc(self, NAdrop=True, onehot=True, fwrite=True,\
                 dataname="new_data.csv", labelsname="new_labels.csv")
Preprocess the raw data for analysis

Drop the NaN and transfer the data into one-hot format. 

Also add index and columns for dataframe  

**<font>Parameters:</font>**  
NAdrop: bool, default = True  
&emsp; If true, it will drop all empty lines for raw data.

onehot: bool, default = True  
&emsp; If true, process the one-hot encoding.(**<font>Must be done before later process<font>**)

fwrite: bool, default = True  
&emsp; If true, write the data into files with specified name

dataname: str  
&emsp; the name of the processed data file (for debuging)

labelsname: str  
&emsp; the name of the label file

    data_split(self, train_size=0.8, shuffle=False)
split the training data into training and testing parts.  
**<font>Parameters:</font>**  
train_size：float  
&emsp; the size of the training data  

shuffle: bool, default = False  
&emsp; if true, shuffle the dataset  

### Variables

    self.data
Processed data
Type: DataFrame
    
    self.labels
Processed labels
Type: DataFrame

    self.train_x, self.train_y, self.test_x, self.test_y
Datasets and labels for training and testing
Type: DataFrame

```python
class DataProcess:
    
    def __init__(self, df_items, df_labels):
        self.rawdata = df_items
        self.rawlabels = df_labels
        
    def Data_preproc(self, NAdrop=True, onehot=True, fwrite=True,\
                     dataname="new_data.csv", labelsname="new_labels.csv"):
        if NAdrop:
            self.rawdata.dropna(axis=0, how='all', inplace=True)
        if onehot:
            self.data = pd.get_dummies(self.rawdata, prefix='').groupby(axis = 1, level = 0).sum()
            self.data.columns = self.data.columns.str.replace("_", "")
            
        self.itemtype = self.data.shape[1]
        self.custnum = self.data.shape[0]
        
        self.data.index = ["Custom No." + str(index) for index in range(self.custnum)]
        self.labels = self.rawlabels.iloc[self.rawdata.index,:]
        self.labeltype = np.unique(self.labels).shape[0]
        self.labels.index = ["Custom No." + str(index) for index in range(self.custnum)]
        
        if fwrite:
            self.data.to_csv(dataname)
            self.labels.to_csv(labelsname)
            
    def Data_split(self, train_size=0.8, shuffle=False):
        if train_size == 1.0:
            train_size = 0.999999
        self.train_x, self.test_x, self.train_y, self.test_y =\
        train_test_split(self.data, self.labels, train_size = train_size, shuffle=shuffle)
    
```

## Class Dict
    class Dict(data)  
class Dict takes the training data point as data.  
**<font>This class can generate two dictionaries mapped from the products to unique number labels(IDs) and reversed. It also provides two functions to transfer the products to IDs and reversed. </font>**  

### Methods
    __init__(self, data)
Initialize self and store maps between IDs and products in dictionaries.  
**<font>Parameters:</font>**  
data: dataframe  
    &emsp;the processed training data. 

    i2p(self, df)
Transfer from IDs to products.  
**<font>Parameters:</font>**  
df: dataframe  
    &emsp;a list containing ids that represents the products

    p2i(self, df)
Transfer from products to IDs.  
**<font>Parameters:</font>**  
df: dataframe  
    &emsp;a list of products
    


```python
class Dict():
    
    def __init__(self, data):
        self.data = data
        self.id_dict = dict(enumerate(data.columns))
        self.prod_dict = dict(zip(data.columns, list(range(len(data.columns)))))
        
    def i2p(self, df):
        return [self.id_dict[i] for i in np.asarray(df).ravel()]
    
    def p2i(self, df):
        return [self.prod_dict[i] for i in np.asarray(df).ravel()]
    
myDict = Dict(myData.data)
```

# 3. Data Exploration and Visualization

    GetPairMap(data)
Get the occurance frequency of each product pairs and store them as an array.  
**<font>Parameters:</font>**  
data: processed training datapoint.  
**<font>Return:</font>**  
pairs: 

type: array

an array containing number of occurance of pairs on their corresponding coordinates.

```python
def GetPairMap(data):
    adata = np.asarray(data)*0.5
    pairs = np.ones([119,119])
    order0 = list(range(119))
    for sft in range(1,119):
        ndata = np.hstack((adata[:,sft:], adata[:,:sft]))
        order1 = list(range(sft,119)) + list(range(sft))
        ndata = ((np.floor(ndata + adata)).sum(axis=0))
        pairs[order0,order1] = ndata
    return np.array(pairs)
```

## Class PlotChart
    class PlotChart(data)
data: processed data.  
**<font>The class Plotchart implements the visualization of training data in various methods. </font>**  

### Methods
    __init__(self, data)
Initialize self.    
    
    Pie(self, sort=True, num=10)
Draw the pie chart of ten most buying products.

    Hist(self, sort=True, num=10)
Show the frequency distribution histogram of ten most buying products.

    Heatmap(self, GetMap=GetPairMap)
Show the heatmap of the pairs bought together.

    CustBuy(self)
Plot the numbers of customers that brought specific number of products.  
For example, only one person bought 18 products.

```python
class PlotChart:
    
    def __init__(self, data):
        self.data = data
        
    def Pie(self, sort=True, num=10):
        if sort:
            mpie = self.data.apply(sum).sort_values(ascending=[False])[:num]
        else:
            mpie = self.data.apply(sum)[:num]
        plt.pie(x=mpie.values, labels=mpie.index)
        plt.show()
        
    def Hist(self, sort=True, num=10):
        if sort:
            number = self.data.apply(sum).sort_values(ascending = False)[:num]
        else:
            number = self.data.apply(sum)[:num]
        plt.bar(list(range(num)), number.values)
        plt.xticks(list(range(num)), number.index, rotation='90')
        plt.show()
        
    def Heatmap(self, GetMap=GetPairMap):
        PairMap=pd.DataFrame(GetPairMap(self.data), columns=self.data.columns, index=self.data.columns)
        print("Calculation finish.")
        print("Drawing...")
        f, ax = plt.subplots(figsize=(30, 30))
        ax = sns.heatmap(PairMap, vmax=305)
        
    def CustBuy(self):
        #This function shows the number of consumers buying different numbers of products
        data_count=self.data.apply(sum,axis=1).sort_values(ascending=[False]).value_counts()
        plt.bar(data_count.index,data_count.values)
        plt.xticks(list(data_count.index), data_count.index)
        plt.show()

myPlot = PlotChart(myData.data)
```

# 4. Mechine Learning: Training

## Class SvmTrain
    class SvmTrain(train_x, train_y)
train_x is an array of training datapoints and train_y is an array of training labels after being translated to IDs.   
**<font>This class implements SVM method to predict the next buying product of given list. </font>**  

### Methods
    __init__(self, train_x, train_y)
Initialize self with training data and labels.

    Train(self, C=100, gamma=0.001)
Implement SVM method to construct the model.  
**<font>Parameters:</font>**  
C: float, default = 100  
    &emsp; Regularization parameter.  
gamma: float, default = 0.001  
    &emsp; Kernel coefficient.  
    (Learn more detail in sklearn.svm.SVC documentaion: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC)
    
    Classifier(self, test_x)
Predict the test result with trained model.  
**<font>Parameters:</font>**  
test_x: dataframe  
    &emsp; A data frame containing test data point.  

**<font>Return:</font>**  
self.pred: list  
    &emsp; A list containing the predict result of product's ids.
    
    Evaluation(self, ans, show=False)
Evaluate the training model.  
**<font>Parameters:</font>**  
ans: list  
    &emsp; A list containing the predict result of product's ids.  
show: bool, default = False  
    &emsp; Show the correct predictions (for debuging)  
 
**<font>Return:</font>**   
accurancy: float  
    &emsp; The accuracy of the model.  

```python
class SvmTrain:
    
    def __init__(self, train_x, train_y):
        #the labels should be translated before using
        self.X = train_x
        self.Y = train_y
    
    def Train(self, C=100, gamma=0.001):
        self.clf = svm.SVC(decision_function_shape='ovo', C=C, gamma=gamma)
        self.clf.fit(self.X, self.Y)
        self.clf.decision_function_shape = "ovr"
        
    def Classifier(self, test_x):
        self.dec = self.clf.decision_function(np.asarray(test_x))
        temp = np.asarray([self.dec.argmax(axis=1)]).ravel()
        self.pred = [np.unique(Y)[i] for i in temp]
        return self.pred
        
    def Evaluation(self, ans, show=False):
        #ans should be translated to id before use
        corr = 0
        for i in range(len(ans)):
            if ans[i] == self.pred[i]:
                corr += 1
                if show:
                    print(myDict.i2p(ans[i]))  
        return corr/len(ans)
        
```

```python
# mechine learning

# transform the dataset into array-type and translate the labels into IDs
X = np.asarray(myData.train_x)
Y = myDict.p2i(np.asarray(myData.train_y).ravel())
# Initialize the SVM module and train the module
mytrain = SvmTrain(train_x=X, train_y=Y)
mytrain.Train()
```

```python
# convert the result into produts name and ouput to csv file.
# here we use the test_x split from the training data to test
# can also be changed to train_x(training data) or test(processed test data)
pd.DataFrame(myDict.i2p(mytrain.Classifier(test_x=np.asarray(myData.test_x)))).to_csv("Prediction.csv")
```

```python
train_x = myData.train_x
test_x = myData.test_x
train_y = pd.get_dummies(myData.train_y)
test_y = pd.get_dummies(myData.test_y)
```

```python
# Get the data

###### GET THE TRAIN TARGET UNIQUE LIST #####
def Get_train_targets(train_y):
    train_y_dummies = pd.get_dummies(list(train_y[0]))

    # get a dictionary
    product_dict = {}
    for i,product in enumerate(myData.data.columns):
        product_dict[product] = i

    #rename the dummies
    train_y_idx = train_y_dummies.rename(columns=product_dict)

    #return the idxmax value
    train_targets = train_y_idx.idxmax(axis=1)
    
    return train_targets

#### GET THE TRAINING DATAPOINT #####
def Get_train_datas(train_x):
    #### receive one hot data!!!! #####
    one_hot_training_data = train_x
    return one_hot_training_data

class CustomDataset():

    def __init__(self, data, labels):

        self.target = torch.LongTensor(labels)
        self.data = torch.Tensor(np.asarray(data))
        
    def __len__(self):
        return self.data.shape[0]

    def __getitem__(self, idx):
        return self.data[idx], self.target[idx]
    
#### BATCH THE DATAPOINT ####
def Get_batch(one_hot_training_data, train_targets, size):
    # zip the data and target
    training_data = CustomDataset(one_hot_training_data,train_targets)
    # batch the data point
    train_loader = torch.utils.data.DataLoader(training_data, batch_size=size, shuffle=True)
    return train_loader
```

```python
train_loader = Get_batch(Get_train_datas(train_x), Get_train_targets(myData.train_y), 30)
test_loader = Get_batch(Get_train_datas(test_x), Get_train_targets(myData.test_y), 30)
```

```python
class Neural_Network(torch.nn.Module):
    
    def __init__(self, input_dim, num_classes):
        
        super(Neural_Network, self).__init__()
        
        self.input_dim = input_dim
        self.num_classes = num_classes
        
        self.layer1 = torch.nn.Linear(self.input_dim, 10)
        
        self.layer2 = torch.nn.Linear(10, self.num_classes)
        
    def forward(self, x):
        
        x = self.layer1(x.view(-1, self.input_dim))
        x = F.sigmoid(x)
        
        x = self.layer2(x)
        
        return x
```

```python
neural_net = Neural_Network(119, 119)
optimizer = torch.optim.SGD(params=neural_net.parameters(), lr=0.01)
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
# training loop:

for epoch in range(20):
    running_loss = 0.0
    
    for i, (x, y) in enumerate(train_loader, 1):
        
        # set optimizer gradients to zero
        optimizer.zero_grad()
        
        # forward pass
        predictions = neural_net.forward(x)
                
        # backward pass + optimization step
        loss = loss_fn(predictions, y)
        loss.backward()
        optimizer.step()
        
        # print statistics
        running_loss += loss.item()
        
        if i % 1000 == 0:
            print(f'Epoch: {epoch}, loss: {running_loss / i}')
        
    print(f'Loss after epoch: {epoch} = {running_loss / len(train_loader)}')
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data, labels in test_loader:
        outputs = neural_net(data)
        _, predicted = torch.max(outputs.data, 1)
        #print(predicted)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print('Accuracy: %d %%' % (
    100 * correct / total))
```
